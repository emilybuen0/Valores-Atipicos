```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
china = pd.read_csv('HongKong_China_Data.csv')
china.info()
```

```python
valores_nulos = china.isnull().sum()
valores_nulos
```

```python
china1 = china.fillna(method = "bfill").fillna(method = "ffill")
china1
```

```python
valores_nulos = china1.isnull().sum()
valores_nulos
```

```python
cconvert = ["host_id"]
china1[cconvert] = china1[cconvert].astype(object)

china1['price'] = china1['price'].str.replace('$', '').str.replace(',', '').astype(float)

china1.info()
```

```python
cuantitativas = china1.select_dtypes(include=['float64', 'int64', 'float', 'int'])
cualitativas = china1.iloc[ : , [0]]

fig = plt.figure(figsize = (15, 8))
cuantitativas.plot(kind = 'box', vert = False)
plt.title("Valores atípicos del DataFrame")
plt.show() #Dibujamos el diagrama
```

```python
#detectar el rango intercualitico 
y = cuantitativas

porcentaje25 = y.quantile(0.25) #Q1
porcentaje75 = y.quantile(0.75) #Q3
iqr = porcentaje75 - porcentaje25

limiter_sup_iqr = porcentaje75 + 1.5*iqr 
limiter_inf_iqr = porcentaje25 - 1.5*iqr 

print(limiter_sup_iqr)
print(limiter_inf_iqr)
```

```python
china3_iqr = cuantitativas[(y<=limiter_sup_iqr)&(y>=limiter_inf_iqr)]
china3_iqr
```

```python
valores_nulos=china3_iqr.isnull().sum()
valores_nulos
```

```python
china4_iqr=china3_iqr.copy()
china4_iqr=china4_iqr.fillna(round(china3_iqr.mean(),1))
china4_iqr
```

```python
China = pd.concat([cualitativas, china4_iqr], axis=1)
China
```

```python
valores_nulos=China.isnull().sum()
valores_nulos
```

```python
China.to_csv("China_NoNulos.csv")
```
